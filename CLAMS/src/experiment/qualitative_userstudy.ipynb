```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.DataFrame({
    "ratio":   np.array([17, 15, 14, 7, 4]) / 18,
    "factors": ["Density", "Shape", "Proximity", "Distribution", "Patterns"]
})


plt.figure(figsize=(8, 3))
sns.set_style("whitegrid")
sns.barplot(x="ratio",y="factors", data=df,)

plt.xlim(0, 1)

plt.tight_layout()
plt.savefig("qualitative_userstudy.pdf", dpi=300)
plt.savefig("qualitative_userstudy.png", dpi=300)
```
